DATA VISUALIZATION

Plot 1: Distribution of Age 

```python
sns.histplot(df['Age'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Sex',data=df,palette='Set2')
plt.title('Count of Gender')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

 Plot 3: Boxplot of Credit Amount by Purpose ---


```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Purpose', y='CreditAmount', data=df)
plt.title('Credit Amount by Purpose')
plt.xlabel('Purpose')
plt.ylabel('Credit Amount')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# --- Plot 4: Correlation Heatmap ---
plt.figure(figsize=(10, 8))
correlation_matrix = df.select_dtypes(include=[np.number]).corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```

Plot 5: Housing Type Count

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Housing', data=df, palette='pastel')
plt.title('Count of Housing Types')
plt.xlabel('Housing')
plt.ylabel('Court')
plt.tight_layout()
plt.show()
```

```python
df['CreditRisk'] = ((df['CreditAmount'] <= 2500) & (df['Duration'] <= 15)).astype(int)
```

```python
df.groupby('CreditRisk')['CreditAmount'].mean().plot(kind='bar', color=['red', 'green'])
plt.title("Average Credit Amount by Credit Risk")
```

```python
df['CreditRisk'] = df['CreditRisk']

age_group_credit_risk = df.groupby(['AgeGroup', 'CreditRisk']).size().unstack()

age_group_credit_risk.plot(kind='bar', stacked=True, color=['red', 'green'], edgecolor='black')
plt.title("Credit Risk Distribution by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.legend(["Bad Credit Risk (0)", "Good Credit Risk (1)"])
plt.tight_layout()
plt.show()
```

```python
sex_credit_risk = df.groupby(['Sex', 'CreditRisk']).size().unstack()

sex_credit_risk.plot(kind='bar', stacked=True, color=['red', 'green'], edgecolor='black')
plt.title("Credit Risk Distribution by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.legend(["Bad Credit Risk (0)", "Good Credit Risk (1)"])
plt.tight_layout()
plt.show()
```
